# Extracting Currency from a String

In this exercise, I was given a column of strings.  I was tasked to extract the currency embedded in each string.

**Solution:**  I created a regular expression identifying the currency format, then applied re.findall within a lambda row function to generate a list of currency strings.  A new column was created using the first value in each list.

```python
import pandas as pd
import re

df = pd.read_csv('Workbook1.csv', names=['String'])
df
```

```python
# Establishing the regular expression for currency
pattern = '\$\d*\.\d{2}'


# findall returns a list, creating a new column with the first value from list.
df['Amount'] = df['String'].apply(lambda x: re.findall(pattern, x)[0])


# Inspecting the dataframe
df
```
